```python
# Importing libraries
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt
import numpy as np
from random import randint
```

```python
# Adjust these values to see difference in graph
num_clusters = 3
num_points = 30
max_random = 1000

# Generating data point for the K means model
X = np.array([[0, 0]])


for j in range(num_points):
    X = np.vstack([X, [randint(0, randint(0, max_random)), randint(0, randint(0, max_random))]])

# The K means model
model = KMeans(n_clusters=num_clusters, random_state=0)
model.fit(X)

# Output
predictions = model.predict(X)
print("predictions:", predictions)
plt.scatter(X.T[0], X.T[1], c=predictions)
```
